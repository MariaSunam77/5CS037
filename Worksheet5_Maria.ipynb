```python
#Cost function is Mean Squared Error(MSE)
def cost_function(X, Y, W):

    n= X.shape[1]  # no.of samples
    Y_pred =np.dot(W.T, X)  # predicted values
    error = Y_pred - Y       # predicted-actual
    cost = (1 /(2* n))*np.sum(error** 2)
    return cost
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    cost_history =[0]*iterations
    n = X.shape[1]# no. of samples

    W_update = W.copy()

    for iteration in range(iterations):
        #s1.Predicted values
        Y_pred = np.dot(W_update.T, X)  # shape: (n,)

        #s2:for loss
        loss = Y_pred - Y  # shape: (n,)

        #s3:for gradient
        dw = (1 / n)*np.dot(X,loss)  # shape: (d,)
        # Step 4: Update weights
        W_update = W_update - alpha * dw

        # s5:for cost
        cost_history[iteration] = cost_function(X, Y, W_update)
    return W_update, cost_history
```

```python
def rmse(Y,Y_pred):
   n = len(Y)
   error = Y-Y_pred
   rmse_value = np.sqrt(np.sum(error ** 2) / n)
   return rmse_value
```

To Do-9

```python
# Model evaluation- R2
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y)** 2) # Total sum of squares
    ss_res = np.sum((Y - Y_pred)** 2)
    r2_value = 1 -(ss_res /ss_tot)
    return r2_value
```

To Do-10

```python
def main():
  # s1: Load the dataset
    data = pd.read_csv('student.csv')

    # s2: Prepare features (X) and target (Y)
    X = data[['Math', 'Reading']].values.T  #transpose to get dxn
    Y = data['Writing'].values              #shape:n

    # s3: Splitting data 70-30
    n = X.shape[1]
    np.random.seed(42)
    indices = np.random.permutation(n)
    train_size = int(0.7*n)
    train_idx = indices[:train_size]
    test_idx = indices[train_size:]

    X_train = X[:, train_idx]
    X_test  = X[:, test_idx]
    Y_train = Y[train_idx]
    Y_test  = Y[test_idx]

    # s4:initializing weights,learning rate,iterations
    d = X_train.shape[0]
    W = np.zeros(d)
    alpha = 0.00001
    iterations = 1000

    # s5:gradient Descent
    W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    # s6: Make prediction test set
    Y_pred = np.dot(W_optimal.T, X_test)

    # s7:evaluate model
    model_rmse = rmse(Y_test, Y_pred)
    model_r2   = r2(Y_test, Y_pred)

    # s8:Output
    print("Final Weights:", W_optimal)
    print("Cost History for first 10 iterations):", cost_history[:10])
    print("RMSE on Test Set:", model_rmse)
    print("R-Squared on Test Set:", model_r2)


# Execute main
if __name__ == "__main__":
    main()

```
