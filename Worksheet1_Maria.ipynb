```python
import numpy as np

A=np.array([[3,4],[7,8]])
B=np.array([[5,3],[2,1]])

A_inv= np.linalg.inv(A)
result1=np.dot(A,A_inv)
print("A.A−1 = I: ",np.allclose( result1,np.eye(2)))
AB= np.dot(A, B)
BA=np.dot( B,A)
print("AB ≠ BA:",not np.array_equal(AB,BA))

AB_T =AB.T
BT_AT =np.dot (B.T, A.T)
print ("(AB)T = BT AT: ",np.array_equal( AB_T,BT_AT))
A_eq=np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B_eq=np.array([-1,-3,9])
solution =np.linalg.solve (A_eq,B_eq)
print("Solution is:", solution)
```
